```python
import pandas as pd
import numpy as np
##from matplotlib import pyplot as plt
import warnings
##import plotly.graph_objects as go

##plt.rcParams['font.family'] = 'SimHei'    #设置中文字体
##plt.rcParams['axes.unicode_minus'] = False   #中文字体状态下负号（-）正常显示
warnings.filterwarnings('ignore') #不显示代码警告（警告不是报错）
np.set_printoptions(suppress=True) #为了直观的显示数字，不采用科学计数法
pd.set_option('display.float_format', lambda x: '%.2f' % x) #设置浮点数显示格式保留两位小数
```

```python
#设置数据源文件路径
file = r'C:\Users\HR\Desktop\预算系统工作低稿\货币资金登记表20210130.xlsx'  #此处为相对路径，可根据budget_data.xlsx文件的实际路径更改此变量值
df01 = pd.read_excel(file,sheet_name='2020.09货币资金登记表-总表') #导入登记表，如果加上index_col=2，就是以二级分类作为索引
df01['摘要']=df01['摘要'].str.replace('，','/')
df01['摘要']=df01['摘要'].str.replace(',','/')
df01=df01.fillna(0)
df01['行为'],df01['对款方']=df01['摘要'].str.split('/',1).str
df02=df01[~df01['摘要'].isin([0])]
df02['新日期']=pd.to_datetime(df02['日期'],format="%Y-%m-%d",errors='coerce')
```

```python
df02=df02[df02['新日期']>=pd.datetime(2021,1,1)]#按日期选出大于2021年1月1日的数据
df02
```

```python
df03=df02.groupby(['二级分类','金额']).sum()
df03
```
