```python
import pandas as pd
import matplotlib.pyplot as plt
file="运费明细表.xlsx"
df01=pd.read_excel(file,sheep_name="运费明细")
df02=pd.pivot_table(df01,index=["出口日期"],values=["应收费用","实际重量"],
                    aggfunc={"应收费用":"sum","实际重量":"sum"})#columns=["账单号码"],fill_value=0, margins=True,margins_name="汇总"
df03=df02.set_index(["应收费用"]).stack().reset_index()            
df03.head(30)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file="运费明细表.xlsx"
df01=pd.read_excel(file,sheep_name="运费明细")
df02=pd.pivot_table(df01,index=["出口日期"],values=["应收费用","实际重量"],columns=["账单号码"],margins=True,margins_name="汇总",
                    aggfunc={"应收费用":"sum","实际重量":"sum"})#columns=["账单号码"],fill_value=0, margins=True,margins_name="汇总"
df02.head(20)
```

```python
with pd.ExcelWriter(file, mode='a',engine='openpyxl') as writer:
    df02.to_excel(writer,sheet_name='001',index=False)  #新开一工作表001而保存文件
```
